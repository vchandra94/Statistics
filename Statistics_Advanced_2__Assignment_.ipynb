**Question-1) What is hypothesis testing in statistics?**

**Answer-1)**

Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It helps determine whether there is enough evidence in the sample to support a specific claim (hypothesis) about the population.

**Key Concepts in Hypothesis Testing:**

1. **Null Hypothesis (H₀):**

    A statement that there is no effect or no difference.

    It represents the default or status quo.

    Example: "The mean test score of students is 75."

2. **Alternative Hypothesis (H₁ or Ha):**

    A statement that contradicts the null hypothesis.

    Represents what you want to test for.

    Example: "The mean test score of students is not 75."

3. **Significance Level (α):**

    The probability threshold for rejecting the null hypothesis.

    Common choices: 0.05 (5%), 0.01 (1%).

    It controls the risk of a Type I error (false positive).

4. **Test Statistic:**

    A calculated value from the sample data used to evaluate the hypothesis.

    Examples: Z-test, t-test, chi-square test.

5. **P-value:**

    The probability of obtaining a result as extreme as, or more extreme than, the observed result, assuming the null hypothesis is true.

    If p-value ≤ α, reject H₀ (evidence supports H₁).

    If p-value > α, fail to reject H₀ (insufficient evidence for H₁).

6. **Decision:**

    Reject H₀: Conclude that there is significant evidence for the alternative hypothesis.

    Fail to Reject H₀: No sufficient evidence to support H₁.

**Question-3) Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

**Answer-3)**

The significance level, often denoted by the symbol α (alpha), is a threshold that defines how strong the evidence must be in order to reject the null hypothesis in hypothesis testing.

**What Is the Significance Level?**

It is the maximum probability of making a Type I error, which means rejecting the null hypothesis when it is actually true.

Common values for the significance level are 0.05 (5%) or 0.01 (1%).

For example, if we choose α = 0.05, it means we are willing to accept a 5% chance of mistakenly rejecting the null hypothesis.

**Role of the Significance Level in Hypothesis Testing:**

1. After calculating the p-value from the sample data, we compare it to the significance level (α).

2. Decision Rule:

    i. If p-value ≤ α → Reject the null hypothesis (H₀).

    There is enough evidence to support the alternative hypothesis (H₁).

    ii. If p-value > α → Fail to reject the null hypothesis (H₀).

    There is not enough evidence to support the alternative hypothesis.



**Example:**

1. Imagine testing whether a new drug is more effective than the current one.

    H₀: The new drug is no more effective than the current one.

    H₁: The new drug is more effective.

1. Suppose after the test, the p-value comes out to be 0.03, and α was set at 0.05.

    Since 0.03 ≤ 0.05, we reject the null hypothesis and conclude that there is significant evidence the new drug is more effective.


**It is Important because ?**

1. The significance level controls how conservative or liberal your test is.

2. A lower α (e.g., 0.01) makes it harder to reject H₀, reducing the chance of false positives but increasing the chance of false negatives (Type II errors).

3. It provides a clear, objective rule for decision-making, rather than relying on subjective judgment.



In simple words, the significance level sets the bar for how surprising the data needs to be before we say "the null hypothesis probably isn’t true." It helps us control the risk of making a wrong decision.

**Question-4) What are Type I and Type II errors? Give examples of each.**

**Answer-4)**

In hypothesis testing, a decision is made about the null hypothesis based on sample data. Since you don't have information about the entire population, there's always a chance of making the wrong decision. These incorrect decisions are categorized into two types of errors.


**1.Type I Error (False Positive):-**

A Type I error occurs when you reject a true null hypothesis. This is also called a "false positive" because your test is incorrectly indicating an effect or difference when one doesn't actually exist. The probability of making a Type I error is set by the significance level (α). A smaller α reduces the risk of this error.



**Example of a Type I Error:**

A medical company tests a new drug, with the null hypothesis being that the drug has no effect on a disease. They conduct a study and, due to a random chance in their sample, the results show a positive effect, so they reject the null hypothesis. They conclude the drug works, but in reality, it's ineffective. This is a Type I error.



**2.Type II Error (False Negative):-**

A Type II error occurs when you fail to reject a false null hypothesis. This is also called a "false negative" because your test is failing to detect an effect or difference that actually exists. The probability of a Type II error is denoted by beta (β).



**Example of a Type II Error:**

A manufacturer's quality control department tests products, with the null hypothesis being that the products meet the quality standard. A batch of products is actually defective, but the test fails to detect the flaws. The department concludes the products are fine and releases them. This is a Type II error—they missed a real problem.

**Question-5) What is the difference between a Z-test and a T-test? Explain when to use each.**

**Answer-5)**

**The key difference between a Z-test and a T-test** lies in what you know about the population's characteristics, specifically the standard deviation, and the size of your sample.

A Z-test is used when you know the population standard deviation (σ) or when you have a large sample size (typically n≥30). With a large sample, the Central Limit Theorem suggests that the sample mean's distribution will be approximately normal, so you can use the standard normal (Z) distribution for your calculations.


A T-test is used when you do not know the population standard deviation and have a small sample size (typically n<30). Because you're estimating the population standard deviation from your small sample, there's more uncertainty. The T-test uses a T-distribution, which has "heavier tails" than the normal distribution. This accounts for the extra variability and uncertainty associated with a smaller sample, making it a more appropriate and conservative test.


***When to Use Each***

**Use a Z-test When:**

The population standard deviation is known.

The sample size is large (n≥30). The Central Limit Theorem allows you to treat the sample distribution as normal, even if the original population isn't.


**Example**:

A company has historical data showing that the average lifespan of a certain light bulb is 1,000 hours with a population standard deviation of 100 hours. A new batch of 50 bulbs is produced, and you want to test if their average lifespan is different. Since you know the population standard deviation and the sample size is large (n=50), a Z-test is appropriate.

**Use a T-test When:**

1. The population standard deviation is unknown.

2. The sample size is small (n<30).

3. The data is approximately normally distributed.

***Example:***

A teacher wants to test if a new teaching method improves student test scores. They apply the method to a class of 20 students and want to compare their average score to the school's historical average. Since they don't know the population standard deviation of all students' scores and the sample size is small, they should use a T-test. The T-test accounts for the greater uncertainty that comes with a smaller group.

```python
"""
Question-7)  Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

#Answer-7)
"""

import numpy as np
from scipy import stats

def z_test_from_sample(sample_data, population_mean, population_std_dev, alpha=0.05):

    # Convert sample data to a numpy array for easier calculations
    sample = np.array(sample_data)

    # Calculate sample statistics
    sample_mean = np.mean(sample)
    n = len(sample)

    print(f"Sample Size (n): {n}")
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Hypothesized Population Mean (μ₀): {population_mean}")
    print(f"Known Population Standard Deviation (σ): {population_std_dev}")
    print("-" * 30)

    # Step 1: State the Hypotheses
    # Null Hypothesis (H₀): The true population mean is equal to the hypothesized mean.
    # H₀: μ = μ₀
    # Alternative Hypothesis (Hₐ): The true population mean is not equal to the hypothesized mean.
    # Hₐ: μ ≠ μ₀
    print("Step 1: Stating the Hypotheses")
    print(f"Null Hypothesis (H₀): The population mean is equal to {population_mean}")
    print(f"Alternative Hypothesis (Hₐ): The population mean is not equal to {population_mean}")
    print("-" * 30)

    # Step 2: Calculate the Z-statistic
    # Z = (sample_mean - population_mean) / (population_std_dev / sqrt(n))
    z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(n))
    print("Step 2: Calculating the Z-statistic")
    print(f"Calculated Z-statistic: {z_score:.4f}")
    print("-" * 30)

    # Step 3: Calculate the p-value
    # For a two-tailed test, the p-value is 2 * P(Z > |z_score|)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    print("Step 3: Calculating the p-value")
    print(f"Calculated p-value: {p_value:.4f}")
    print("-" * 30)

    # Step 4: Make a Decision
    print(f"Step 4: Making a Decision (using α = {alpha})")
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis.")
        print(f"The p-value ({p_value:.4f}) is less than the significance level ({alpha}).")
        print("There is sufficient evidence to conclude that the sample mean is statistically different from the hypothesized population mean.")
    else:
        print("Conclusion: Fail to reject the null hypothesis.")
        print(f"The p-value ({p_value:.4f}) is greater than the significance level ({alpha}).")
        print("There is not enough evidence to conclude that the sample mean is statistically different from the hypothesized population mean.")

# Sample data provided by the user
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Assume the hypothesized population mean (μ) is 50 and the known
# population standard deviation (σ) is 1.0. These values are necessary
# for a Z-test and are based on a typical scenario where this test is used.
population_mean = 50.0
population_std_dev = 1.0

# Run the Z-test analysis
z_test_from_sample(sample_data, population_mean, population_std_dev)
```

```python
#Question-08) Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

#Answer-08)


import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_confidence_interval(data, confidence=0.95):

    sample_mean = np.mean(data)
    sample_std_err = stats.sem(data)  # Standard error of the mean
    degrees_freedom = len(data) - 1

    # Use the t-distribution to calculate the margin of error
    margin_error = sample_std_err * stats.t.ppf((1 + confidence) / 2, degrees_freedom)

    # Calculate the confidence interval
    conf_interval = (sample_mean - margin_error, sample_mean + margin_error)

    return sample_mean, conf_interval

def plot_data_with_confidence_interval(data, sample_mean, conf_interval, population_mean):

    plt.figure(figsize=(10, 6))

    # Plot the histogram of the data
    plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')

    # Plot the true population mean
    plt.axvline(population_mean, color='red', linestyle='--', linewidth=2, label='True Population Mean')

    # Plot the sample mean
    plt.axvline(sample_mean, color='green', linestyle='-', linewidth=2, label='Sample Mean')

    # Plot the confidence interval as a shaded area
    plt.axvspan(conf_interval[0], conf_interval[1], color='gray', alpha=0.3, label='95% Confidence Interval')

    # Add labels and title
    plt.title('Simulated Normal Distribution with Confidence Interval')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# --- Main simulation and plotting logic ---
# Set parameters for the normal distribution
population_mean = 50
population_std_dev = 5
sample_size = 100

# Step 1: Simulate data from a normal distribution
simulated_data = np.random.normal(loc=population_mean, scale=population_std_dev, size=sample_size)

# Step 2: Calculate the 95% confidence interval for the sample mean
sample_mean, conf_interval = calculate_confidence_interval(simulated_data)

# Step 3: Print the results
print(f"Simulated Sample Size: {sample_size}")
print(f"True Population Mean: {population_mean}")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({conf_interval[0]:.2f}, {conf_interval[1]:.2f})")

# Step 4: Plot the data and the confidence interval
plot_data_with_confidence_interval(simulated_data, sample_mean, conf_interval, population_mean)
```

**Question-9-(Part-2): Explain what the Z-scores represent in terms of standard deviations from the mean.**

**Answer9-9- part-2**

A Z-score is a measure of a data point's position relative to the mean of a dataset. It tells you exactly how many standard deviations away from the mean a particular data point is.

1. A positive Z-score means the data point is above the mean. For example, a Z-score of +1 means the value is exactly one standard deviation above the mean.

2. A negative Z-score means the data point is below the mean. A Z-score of -2 means the value is exactly two standard deviations below the mean.

3. A Z-score of 0 indicates the data point is identical to the mean.

The main purpose of a Z-score is to standardize data from different distributions. By converting all values to Z-scores, you can compare a data point from one set to a data point from another, regardless of their original units or scales. This also makes it easy to identify outliers, as a data point with an absolute Z-score greater than 2 or 3 is generally considered unusual.
